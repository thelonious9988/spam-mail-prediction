```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
ds = pd.read_csv('mail_data.csv')
```

```python
ds
```

```python
ds.info()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
x = ds['Message']
y = ds['Category']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, stratify = y, random_state = 1)
```

```python
x_train
```

```python
transform = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase = True)
```

```python
x_train_trans = transform.fit_transform(x_train)
x_test_trans = transform.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train_trans, y_train)
```

```python
train_pred = model.predict(x_train_trans)
train_acc = accuracy_score(train_pred, y_train)
print('Accuracy is : ', train_acc)
```

```python
test_pred = model.predict(x_test_trans)
test_acc = accuracy_score(test_pred, y_test)
print('Accuracy is : ', test_acc)
```
